```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv


df = pd.read_csv('../Data/Pokemon.csv')
df.head()
df.info()
df.describe()
df.isnull().sum()
```

```python
# Use bfill directly
df_filled = df.copy()
df_filled['Type 2'] = df_filled['Type 2'].bfill()

# Verify the changes
df_filled.isnull().sum()
```

```python
df_filled['Battle_Strength'] = (df_filled['Attack'] * 0.4 +
                                df_filled['Defense'] * 0.3 +
                                df_filled['Speed'] * 0.2 +
                                df_filled['HP'] * 0.1)


df_filled[['Name', 'Attack', 'Defense', 'Speed', 'HP', 'Battle_Strength']].head()
df_filled.to_csv('../Data/Cleaned_pokemon_data.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df_filled['Battle_Strength'], kde=True)
plt.title('Distribution of Battle Strength')
plt.show()
```
