```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import csv


df = pd.read_csv('../Data/Pokemon.csv')
df.head()
df.info()
df.describe()
df.isnull().sum()
```

```python
# Visualize the distribution of the Attack stat
sns.histplot(df['Attack'], kde=True)
plt.title('Distribution of Attack Stat')
plt.xlabel('Attack')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Attack', y='Defense', data=df)
plt.title('Distribution of Attack Stat')
plt.xlabel('Attack')
plt.ylabel('Frequency')
plt.show()
```
